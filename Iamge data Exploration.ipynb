```python
import glob
import os

folder_path = "/Users/munseunghyeon/input/food-101/food-101/images"
folder_result = glob.glob(f"{folder_path}/*")

file_size_list = []
file_info_list = []
img_num = {}

for i in folder_result :
    result = glob.glob(f"{i}/*")
    img_num.update({i : len(result)})
    for j in result :
        image = cv2.imread(j)
        img_size = os.path.getsize(j)
        height, width, shape = image.shape
        file_info_list.append([height, width, shape])
        file_size_list.append(img_size)
        
```

```python
!python -c "import matplotlib; print(matplotlib.rcsetup.all_backends)"
```

```python
import tkinter
import matplotlib
import matplotlib.pyplot as plt
matplotlib.use('TkAgg')
fig, ax = plt.subplots()
plt.boxplot(file_size_list)

ax.set_title("이미지 사이즈")
ax.set_xlabel("data")
ax.set_ylabel("value")

plt.show()
```

```python
avg_height = np.array(file_info_list)[:,0].mean()
avg_width = np.array(file_info_list)[:,1].mean()
avg_shape = np.array(file_info_list)[:,2].mean()
```

```python
print(avg_height, avg_width, avg_shape)
```

```python
import matplotlib.pyplot as plt
matplotlib.use('TkAgg')


fig, ax = plt.subplots()
ax.bar(["avg_height", "avg_width", "avg_shape"], [avg_height, avg_width, avg_shape])
ax.legend()
plt.show()
```
